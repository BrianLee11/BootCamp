## Answer 1

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
url = "https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

# Convert the 'hour_beginning' column to datetime format
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')

# Extract the day of the week
df['day_of_week'] = df['hour_beginning'].dt.day_name()

# Define weekdays
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

# Filter for weekdays only
df_weekdays_only = df[df['day_of_week'].isin(weekdays)]

# Group by day of the week and sum the pedestrian counts
df_weekday_pedestrians = df_weekdays_only.groupby('day_of_week')['Pedestrians'].sum()

# Sort by weekday order
df_weekday_pedestrians = df_weekday_pedestrians.reindex(weekdays)

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(df_weekday_pedestrians.index, df_weekday_pedestrians.values, marker='o', linestyle='-')

# Add data labels
for i, value in enumerate(df_weekday_pedestrians.values):
    plt.text(i, value, f"{value:,}", ha='center', va='bottom', fontsize=10)

# Labels and title
plt.xlabel("Day of the Week")
plt.ylabel("Total Pedestrian Count")
plt.title("Pedestrian Counts by Weekday")
plt.grid(True)

# Show the plot
plt.plot(df_weekday_pedestrians.index, df_weekday_pedestrians.values, marker='o', linestyle='-')
```

# Question 2: converting non-numerical feature and create correlation spot

- ⁠Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year. Sort the pedestrian count data by weather summary to identify any correlations( with a correlation matrix) between weather patterns and pedestrian counts for the selected year.

- This question requires you to show the relationship between a numerical feature(Pedestrians) and a **non-numerical feature(Weather Summary)**. In such instances we use Encoding. Each weather condition can be encoded as numbers( 0,1,2..). This technique is called One-hot encoding.

- Correlation matrices may not always be the most suitable visualization method for relationships involving categorical data points, nonetheless this was given as a question to help you understand the concept better.


## brainstorming

```python
# Read the dataset
url = "https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

df['location'].unique()
```

```python
# Convert the 'hour_beginning' column to datetime format
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')

df['hour_beginning'].dt.year.unique()
```

```python
# Consider only year = 2019 data
df_2019 = df[df['hour_beginning'].dt.year == 2019]
```

```python
# give a snapshot of the year 2019 data
df_2019.info()
```

```python
# describe the year 2019 data
df_2019.describe()
```

```python
df_2019
```

```python
# Select relevant columns for correlation analysis
correlation_columns = ['Pedestrians', 'weather_summary','temperature', 'precipitation']
df_corr = df_2019[correlation_columns].dropna()

# Convert categorical weather_summary into numerical format using one-hot encoding
df_corr = pd.get_dummies(df_corr, columns=['weather_summary'])
```

```python
# Compute the correlation matrix
correlation_matrix = df_corr.corr()
```

```python
# Display the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Weather vs. Pedestrian Counts (2019)")
plt.show()
```

```python
# Read the dataset
url = "https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

# Convert the 'hour_beginning' column to datetime format
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')

# Consider only year = 2019 data
df_2019 = df[df['hour_beginning'].dt.year == 2019]

# Select relevant columns for correlation analysis
correlation_columns = ['Pedestrians', 'weather_summary','temperature', 'precipitation']
df_corr = df_2019[correlation_columns].dropna()

# Convert categorical weather_summary into numerical format using one-hot encoding
df_corr = pd.get_dummies(df_corr, columns=['weather_summary'])

# Compute the correlation matrix
correlation_matrix = df_corr.corr()

# Display the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Weather vs. Pedestrian Counts (2019)")
plt.show()
```

```python
# Read the dataset
url = "https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

df['hour_beginning'] = pd.to_datetime(df["hour_beginning"], format="%m/%d/%Y %I:%M:%S %p")

df['time_of_day'] = "NAH"

def categorize_time_of_day(number):
    if 5 <= number <= 10:
        return "Morning"
    elif 11 <= number <= 17:
        return "Afternoon"
    elif 18 <= number <= 20:
        return "Evening"
    elif 21 <= number <= 24 or 0 <= number <= 4:
        return "Night"

df['time_of_day'] = "NAH"

for index, row in df.iterrows():  # Iterate over rows
    df.loc[index, 'time_of_day'] = categorize_time_of_day(row['hour_beginning'].hour)



columns_pedestrians = ["Pedestrians", "Towards Manhattan", "Towards Brooklyn"]

df_pedestrians_time_of_day_sum = df.groupby('time_of_day')[columns_pedestrians].sum()

df_pedestrians_time_of_day_mean = df.groupby('time_of_day')[columns_pedestrians].mean()

df_pedestrians_time_of_day_median = df.groupby('time_of_day')[columns_pedestrians].median()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot pedestrian flow (Manhattan vs. Brooklyn) by time of day
df_pedestrians_time_of_day_sum[['Pedestrians','Towards Manhattan', 'Towards Brooklyn']].plot(
    kind='bar', stacked=False, figsize=(10,6)
)

# Labels and title
plt.title("Pedestrian Total Counts by Time of Day (Total, Manhattan vs. Brooklyn)")
plt.xlabel("Time of Day")
plt.ylabel("Total Pedestrian Count")
plt.xticks(rotation=0)
plt.legend(title="Direction")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot pedestrian flow (Manhattan vs. Brooklyn) by time of day
df_pedestrians_time_of_day_mean[['Pedestrians','Towards Manhattan', 'Towards Brooklyn']].plot(
    kind='bar', stacked=False, figsize=(10,6)
)

# Labels and title
plt.title("Pedestrian Mean Counts by Time of Day (Total, Manhattan vs. Brooklyn)")
plt.xlabel("Time of Day")
plt.ylabel("Mean Pedestrian Count")
plt.xticks(rotation=0)
plt.legend(title="Direction")

# Show the plot
plt.show()
```
